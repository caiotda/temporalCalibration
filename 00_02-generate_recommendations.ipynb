```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from  caserec.recommenders.rating_prediction.svdplusplus import SVDPlusPlus
```

```python
movielens = pd.read_csv("datasets/ml20m_filtered.csv")[["userId", "movieId", "genres", "timestamp", "rating"]]
```

```python
movielens
```

```python
X = movielens.drop(columns=['timestamp', 'genres'])
```

```python
X
```

```python
y = movielens[["rating"]]
```

```python
movielens
```

```python
train_df, test_df , _, _ = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train_df
```

```python
train_df.to_csv("datasets/movielens/train.dat", index=False, header=False, sep='\t')
test_df.to_csv("datasets/movielens/test.dat", index=False, header=False, sep='\t')
```

```python
SVDPlusPlus(
        train_file='datasets/movielens/train.dat',
        test_file='datasets/movielens/test.dat',
        output_file='datasets/predictions/svdpp_movielens.dat',
        epochs=20,
        learn_rate=0.005,
        delta=0.02,
        factors=20,
        random_seed=42,
        ).compute()
```
